###  Question #4: What products are most often sold together?

```python
data.head()
```

```python
orders_multiple= data[data['Order ID'].duplicated(keep= False)]
```

```python
orders_multiple['Grouped'] = orders_multiple.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
orders_multiple.head()
```

```python
orders_multiple = orders_multiple[['Order ID', 'Grouped']].drop_duplicates()
```

```python
orders_multiple.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in orders_multiple['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key,': ',value)
```

#### Items most commonly bought together are:
        ('iPhone', 'Lightning Charging Cable') :  1005
        ('Google Phone', 'USB-C Charging Cable') :  987

### Question #5: What product sold the most? Why do you think it did?

```python
quantity_by_product =data.groupby('Product').sum()[['Quantity Ordered']].reset_index()
```

```python
prices = data.groupby('Product').mean()['Price Each'].reset_index()
```

```python
# Group by 'Product' and calculate the mean of 'Price Each'
prices = data.groupby('Product', as_index=False)['Price Each'].mean()

# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(prices['Product'], prices['Price Each'], color='green')
plt.title('Average Price of Each Product')
plt.xlabel('Product')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize = (10,10))

sns.barplot(data = quantity_by_product, x= 'Product', y = 'Quantity Ordered', ax = ax1,color= 'orange')

ax2 =ax1.twinx()

sns.lineplot(data=prices, x = 'Product', y = 'Price Each', ax =ax2)

ax1.set_xticklabels(prices['Product'], rotation ='vertical', size = 10)
```

```python
# Group by 'Product' and sum the 'Quantity Ordered' column
quantity_by_product = data.groupby('Product', as_index=False)['Quantity Ordered'].sum()

# Group by 'Product' and calculate the mean of 'Price Each'
prices = data.groupby('Product', as_index=False)['Price Each'].mean()

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

sns.barplot(data=quantity_by_product, x='Product', y='Quantity Ordered', ax=ax1, color='orange')
ax2 = ax1.twinx()
sns.lineplot(data=prices, x='Product', y='Price Each', ax=ax2)

ax1.set_xticklabels(prices['Product'], rotation='vertical', size=10)
ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='orange')
ax2.set_ylabel('Price Each')

plt.title('Quantity Ordered and Price Each by Product')
plt.show()
# This code will generate a bar pleot for the quantity ordered and a line plot for the average price each product, aligned by their respective product names.







```
