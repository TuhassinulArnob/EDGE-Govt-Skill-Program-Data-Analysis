# Pandas

```python
import pandas as pd
```

```python
# creating a series from a list
data = [10,20,30,40]
s=pd.Series(data,index=['a','b','c','d'])

print(s)

# Accessing Elements by index
print(s['b'])

# performing operations
print(s*2)
```
